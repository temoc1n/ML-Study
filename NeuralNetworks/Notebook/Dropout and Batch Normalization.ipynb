```python
from tensorflow.python.keras import Sequential, layers
from tensorflow.python.keras.callbacks import EarlyStopping
import tensorflow as tf
import numpy as np
import pandas as pd
```

```python
model = Sequential([
    layers.Dense(units=1024, input_shape=[11], activation='relu'),
    layers.Dropout(rate=0.3),
    tf.keras.layers.BatchNormalization(),
    layers.Dense(units=1024, activation='relu'),
    layers.Dropout(rate=0.3),
    tf.keras.layers.BatchNormalization(),
    layers.Dense(units=500, activation='sigmoid'),
    layers.Dropout(rate=0.5),
    tf.keras.layers.BatchNormalization(),
    layers.Dense(units=1)
])
```

```python
model.compile(loss='mae', optimizer='adam')
```

```python
X_train = np.random.rand(1, 11)
Y_train = np.array([1])
```

```python
early_stopping = EarlyStopping(
    restore_best_weights=True,
    patience=20,
    min_delta=0.001
)
```

```python
history = model.fit(X_train, Y_train, epochs=150, batch_size=36, callbacks=[early_stopping])
```

```python
dframe = pd.DataFrame(history.history)
dframe['loss'].plot()
```

```python
X = np.random.rand(1,11)
print(model.predict(X))
print(X)
```
