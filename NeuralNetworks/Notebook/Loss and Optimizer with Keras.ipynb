```python
from tensorflow import keras
from tensorflow.python.keras import Sequential, layers
import tensorflow as tf
import pandas as pd
import numpy as np
```

```python
module = Sequential([
    layers.Dense(units=5, input_shape=[11], activation='relu'),
    layers.Dense(units=10, activation='relu'),
    layers.Dense(units=10, activation='relu'),
    layers.Dense(units=3, activation='relu'),
    layers.Dense(units=1)
])
```

```python
module.compile(loss='mae', optimizer='adam', metrics = ['accuracy'])
```

```python
x = np.random.rand(1,11) * 100
y = np.array([[1]])
```

```python
history = module.fit(x, y, epochs=200, batch_size=36)
```

```python
Y = module.predict(x)
print(y)
print(Y)
```

```python
data_frame = pd.DataFrame(history.history)
data_frame['loss'].plot()
```
