```python
import tensorflow as tf
import numpy as np
import pandas as pd
from tensorflow import keras
from tensorflow.python.keras import Sequential, layers
from tensorflow.python.keras.callbacks import EarlyStopping
```

```python
X_train = np.random.rand(1,5)
Y_train = np.array([1])
input_shape = np.array(X_train.shape)
```

```python
model = Sequential([
    tf.keras.layers.BatchNormalization(input_shape=input_shape),
    layers.Dense(units=1024, activation='relu'),
    layers.Dropout(rate=0.3),
    tf.keras.layers.BatchNormalization(),
    layers.Dense(units=780, activation='relu'),
    layers.Dropout(rate=0.4),
    tf.keras.layers.BatchNormalization(),
    layers.Dense(units=1, activation='sigmoid')
])
```

```python
early_stopping = EarlyStopping(
    min_delta=0.001,
    patience=20,
    restore_best_weights=True
)
```

```python
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['binary_accuracy'])
```

```python
history = model.fit(X_train, Y_train, epochs=230, batch_size=36, callbacks=[early_stopping])
```

```python
dframe = pd.DataFrame(history.history)
dframe['loss'].plot(title="Binary Crossentropy")
```
