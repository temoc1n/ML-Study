```python
from tensorflow import keras
from tensorflow.python.keras.callbacks import EarlyStopping
from tensorflow.python.keras import Sequential, layers
import numpy as np
import pandas as pd
```

```python
model = Sequential([
    layers.Dense(units=36, input_shape=[10], activation='relu'),
    layers.Dense(units=36, activation='relu'),
    layers.Dense(units=1)
])
```

```python
model.compile(loss='mae', optimizer='adam')
```

```python
early_stopping = EarlyStopping(
    min_delta = 0.001,
    patience = 20,
    restore_best_weights = True
)
```

```python
X = np.random.rand(1, 10) * 10
y = np.array([[4]])
```

```python
history = model.fit(
    X, y,
    epochs=100,
    batch_size=36,
    callbacks=[early_stopping]
)
```

```python
Y = model.predict(X)

dframe = pd.DataFrame(history.history)

dframe['loss'].plot()
```
